## **HIERARCHICAL CLUSTERING**

#### **Python Libraries**

```python
import numpy as np
import pandas as pd 
import seaborn as sns
from termcolor import colored
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from yellowbrick.cluster import KElbowVisualizer
from mpl_toolkits.mplot3d import Axes3D

import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 
warnings.filterwarnings("ignore", category=FutureWarning)
```

#### **Dataset**

```python
usarests = pd.read_csv(r'C:\Users\AleynaCihangir\Desktop\MarketPlace PowerBrands Projects\USArrests.csv', index_col = 0)
df = usarests.copy()
df.head()
```

```python
hc_complete = linkage(df, "complete")
hc_average = linkage(df, "average")
```

```python
df.describe().T
```

```python
plt.figure(figsize = (15,5))
plt.title("Hierarchical Clustering Dendogram", fontsize=15, color="blue")
plt.xlabel("Observation Units" , fontsize=12, color="blue")
plt.ylabel("Distances" , fontsize=12, color="blue")
dendrogram(hc_complete,
          leaf_font_size = 10);
```

```python
plt.figure(figsize = (15,5))
plt.title("Hierarchical Clustering Dendogram", fontsize=15, color="blue")
plt.xlabel("Observation Units" , fontsize=12, color="blue")
plt.ylabel("Distances" , fontsize=12, color="blue")
dendrogram(hc_average,
          leaf_font_size = 10);
```

```python
plt.figure(figsize = (15,5))
plt.title("Hierarchical Clustering Dendogram", fontsize=15, color="blue")
plt.xlabel("Observation Units" , fontsize=12, color="blue")
plt.ylabel("Distances" , fontsize=12, color="blue")
dendrogram(hc_complete,
           truncate_mode = "lastp",
           p = 10,
           show_contracted = True,
          leaf_font_size = 10);
```
