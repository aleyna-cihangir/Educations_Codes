##### **PYTHON LIBRARIES**

```python
from termcolor import colored
import numpy as np 
import pandas as pd
import seaborn as sns
```

##### **PANDAS**

```python
seri = pd.Series([1, 20, 3, -12, 5], index=['a', 'b', 'c', 'd', 'e'])
print(colored("\033[1m" + "A Sample Seri is: ", "blue"),'\n', seri) 
```

```python
data = {'Name': ['Ozge', 'Ahmet', 'Sirin', 'Mehmet'],
        'Experience': [3, 0.5, 2.5, 1],
        'EarnMoney': [9000, 2500, 6500, 5000]}
df = pd.DataFrame(data, columns=["Name",'Experience', "EarnMoney"])
print(colored("\033[1m" + "A Sample DataFrame is: ", "blue"),'\n', df, '\n') 
df = sns.load_dataset("titanic")
df
```

```python
print(colored("\033[1m" + "Titanic Dataset Shape: ", "blue"), df.shape) 
print(colored("\033[1m" + "Titanic Dataset Size: ", "blue"), df.size) 
print(colored("\033[1m" + "Titanic Dataset mxn: ", "blue"), df.ndim) 
print(colored("\033[1m" + "Titanic Dataset Columns: ", "blue"), df.columns) 
print(colored("\033[1m" + "Titanic Dataset Types: ", "blue"), '\n',df.dtypes) 
print(colored("\033[1m" + "Titanic Dataset Count NaN Values: ", "blue"), '\n',df.isnull().sum()) 
```

```python
print(colored("\033[1m" + "Titanic Dataset Correlation: ", "blue")) 
df.corr()
```

```python
print(colored("\033[1m" + "3. Rows Age values using LOC: ", "blue"), df.loc[3, 'age'])
print(colored("\033[1m" + "3. Rows Age values using ILOC: ", "blue"), df.iloc[3, 3])
```
